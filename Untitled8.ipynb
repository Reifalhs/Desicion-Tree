```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.impute import SimpleImputer
import joblib
from imblearn.over_sampling import SMOTE

# Baca ulang dataset
data = pd.read_csv('water_potability.csv')

# Tampilkan ringkasan statistik
print(data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot distribusi beberapa fitur
plt.figure(figsize=(12, 8))
sns.boxplot(data=data[['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']])
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Pisahkan fitur dan target
features = data[['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']]
target = data['Potability']

# Tangani nilai NaN dengan mengisi nilai yang hilang
imputer = SimpleImputer(strategy='mean')
features = imputer.fit_transform(features)

# Normalisasi data
scaler = StandardScaler()
features = scaler.fit_transform(features)

# Bagi data menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import joblib

# Latih model Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluasi model
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('Accuracy:', accuracy_score(y_test, y_pred))

# Simpan model, imputer, dan scaler
joblib.dump(model, 'random_forest_model.pkl')
joblib.dump(imputer, 'imputer.pkl')
joblib.dump(scaler, 'scaler.pkl')
```

```python
def predict_water_quality(ph, hardness, solids, chloramines, sulfate, conductivity, organic_carbon, trihalomethanes, turbidity):
    # Load model, imputer, dan scaler
    model = joblib.load('random_forest_model.pkl')
    imputer = joblib.load('imputer.pkl')
    scaler = joblib.load('scaler.pkl')

    # Buat dataframe dari input
    input_data = pd.DataFrame([[ph, hardness, solids, chloramines, sulfate, conductivity, organic_carbon, trihalomethanes, turbidity]],
                              columns=['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity'])

    # Tangani nilai NaN dalam input
    input_data = imputer.transform(input_data)

    # Normalisasi input
    input_data = scaler.transform(input_data)

    # Lakukan prediksi
    prediction = model.predict(input_data)

    # Kembalikan hasil prediksi
    if prediction[0] == 1:
        return "Air Minum Layak Konsumsi"
    else:
        return "Air Minum Tidak Layak Konsumsi"

# Contoh penggunaan fungsi prediksi dengan nilai parameter yang layak konsumsi
result = predict_water_quality(7.0, 150.0, 10000.0, 7.0, 300.0, 400.0, 10.0, 80.0, 3.0)
# result = predict_water_quality(7.5, 100.0, 200.0, 1.0, 50.0, 500.0, 2.0, 60.0, 0.5)
print(result)
```
